```python
import os
import numpy as np
import skimage.io
from cellpose import models, core  

input_dir = r'D:\biyesheji\cells\downsampled_data-20241101T003120Z-001\downsampled_data'

output_dir = os.path.join(input_dir, 'Masks')

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
else:
    print("Existing Mask Directory found. Deleting it and creating a new one.")
    import shutil
    shutil.rmtree(output_dir)
    os.makedirs(output_dir)

image_extensions = ['.tif', '.tiff', '.TIF', '.TIFF']  

all_files = os.listdir(input_dir)
print("Files in the directory:", all_files)

files = [f for f in all_files if os.path.splitext(f)[1] in image_extensions]

if len(files) == 0:
    print("No TIFF images found in the input directory.")
else:
    print(f"Found {len(files)} TIFF images.")

use_GPU = core.use_gpu()
print("Using GPU." if use_GPU else "Using CPU.")

model_type = 'cyto'  
model = models.CellposeModel(gpu=use_GPU, model_type=model_type)


channels = [0, 0]  # [cytoplasm_channel, nucleus_channel]

diameter = 0  
flow_threshold = 0.4 
cellprob_threshold = 0.0 

if len(files) == 0:
    print("No images to process. Exiting the script.")
else:
    print("Starting segmentation...")

    # Process each image
    for filename in files:
        filepath = os.path.join(input_dir, filename)
        print(f"Processing {filename}...")
        
        # Read the image
        img = skimage.io.imread(filepath)
        
        # Run Cellpose segmentation
        masks, flows, styles = model.eval(
            img, 
            diameter=diameter, 
            flow_threshold=flow_threshold,
            cellprob_threshold=cellprob_threshold, 
            channels=channels
        )
        
        # Save the mask as a 16-bit TIFF image
        mask_path = os.path.join(output_dir, 'mask_' + filename)
        skimage.io.imsave(mask_path, masks.astype(np.uint16), check_contrast=False)
        print(f"Saved mask to {mask_path}")
    
    print("Segmentation complete.")
```
